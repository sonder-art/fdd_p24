# Principios del Prompting


## Instruciones claras y especificas

### Usar delimitadores

Resume en 1 parrafo el siguiente texto:  

"
Entiende el modelo: Antes de comenzar, es importante tener una comprensión básica del modelo de lenguaje con el que estás interactuando. ¿Qué tipo de tareas puede realizar? ¿Cuáles son sus limitaciones? Saber qué esperar te ayudará a formular el prompting adecuado.

Especifica el contexto: Al iniciar la conversación con ChatGPT, proporciona un contexto claro para que el modelo sepa sobre qué tema o situación está conversando. Esto puede ser tan simple como una oración introductoria que describa el tema o una pregunta inicial.

Formula preguntas directas: Si deseas respuestas específicas, formula preguntas directas y específicas. Cuanto más claro sea tu interrogante, más probable será que obtengas una respuesta precisa.

Utiliza instrucciones claras: Si deseas que ChatGPT realice una tarea particular, proporciona instrucciones claras y específicas. Por ejemplo, si deseas que resuma un texto, puedes decir: "Por favor, resume el siguiente párrafo en dos oraciones".

Controla la salida con temperaturas: La "temperatura" controla la aleatoriedad en las respuestas del modelo. Un valor bajo (por ejemplo, 0.2) producirá respuestas más determinísticas y conservadoras, mientras que un valor alto (por ejemplo, 0.8) generará respuestas más diversas y creativas.

Limita la longitud de la respuesta: Si deseas respuestas más concisas, puedes establecer un límite de tokens para controlar la longitud de la salida.

Experimenta y ajusta: No te limites a un solo prompt. Experimenta con diferentes enfoques, preguntas y estilos para ver qué tipo de resultados obtienes. Ajusta tus instrucciones según sea necesario para mejorar la calidad de las respuestas.

Revisa y edita: Las respuestas generadas por ChatGPT pueden no ser siempre perfectas. Siempre es una buena práctica revisar y editar las respuestas según tus necesidades o para mejorar la coherencia y precisión.

Mantén la seguridad en mente: Aunque ChatGPT es útil, recuerda que es solo un modelo de lenguaje y puede producir información incorrecta o sesgada. No utilices el modelo para difundir información engañosa o dañina.

Respeto y ética: Utiliza el modelo de manera respetuosa y ética. Evita solicitar contenido inapropiado o insensible, y recuerda que el modelo debe ser tratado como una herramienta que pueda brindar asistencia útil y constructiva.

Siguiendo estos consejos, podrás aprovechar al máximo tu interacción con ChatGPT y obtener resultados más relevantes y precisos. ¡Diviértete experimentando y aprendiendo!
"
